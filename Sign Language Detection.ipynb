```python
import cv2
import os
import time
import uuid
import os
```

```python
IMAGES_PATH = r"C:\Users\zarak\Documents"
```

```python
labels = ['bye', 'welcome', 'yeah', 'nope', 'thankyou']
number_imgs = 15
```

```python
for label in labels:
    os.mkdir(os.path.join(IMAGES_PATH,label))
    cap = cv2.VideoCapture(0)
    time.sleep(5)
    print('Collecting Images for {}'.format(label))
    # time.sleep(5)
    for imgnum in range(number_imgs):
        ret, frame = cap.read()
        imgname = os.path.join(IMAGES_PATH,label, label+'.'+'{}.jpg'.format(str(uuid.uuid1())))
        a = cv2.imwrite(imgname, frame)
        print(a)
        print("iMAGE Save at location {}".format(imgname))
        cv2.imshow('frame', frame)
        time.sleep(5)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    cap.release()
```
